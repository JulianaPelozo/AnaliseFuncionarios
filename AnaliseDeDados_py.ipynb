```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
plt.hist(df['satisfaction_level'], bins=20, edgecolor='black')
plt.title('Distribuição do Nível de Satisfação')
plt.xlabel('Nível de Satisfação')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(df['satisfaction_level'], shade=True, color="pink")
plt.title('Curva de Densidade/Nível de Satisfação')
plt.xlabel('Nível de Satisfação')
plt.ylabel('Densidade')
plt.show()
```

```python
plt.figure(figsize=(8,6))


sns.kdeplot(df[df['left'] == 0]['satisfaction_level'],
            shade=True, color="green", label="Ficaram")


sns.kdeplot(df[df['left'] == 1]['satisfaction_level'],
            shade=True, color="red", label="Saíram")

plt.title('Funcionários - Saíram vs Ficaram')
plt.xlabel('Nível de Satisfação')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['satisfaction_level'], showfliers=True)
plt.title('Boxplot/Nível de Satisfação')
plt.xlabel('Nível de Satisfação')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=df['satisfaction_level'], showfliers=False)
plt.title('Boxplot sem outliers')
plt.xlabel('Nível de Satisfação')
plt.show()
```
