```python
plt.figure(figsize=(8,6))


sns.kdeplot(df[df['left'] == 0]['satisfaction_level'],
            shade=True, color="green", label="Ficaram")


sns.kdeplot(df[df['left'] == 1]['satisfaction_level'],
            shade=True, color="red", label="Sa√≠ram")

plt.title('Funcion√°rios - Sa√≠ram vs Ficaram')
plt.xlabel('N√≠vel de Satisfa√ß√£o')
plt.ylabel('Densidade')
plt.legend()
plt.show()
```
